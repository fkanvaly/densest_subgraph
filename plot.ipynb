```python
import numpy as np
import pandas as pd
from pathlib import Path

from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.palettes import Category10
from bokeh.transform import factor_cmap

import bokeh.io
bokeh.io.output_notebook()
```

```python
p = Path('../output/densest_subgraph/')
files = list(p.glob('*.csv'))
files
```

```python
df = pd.concat([pd.read_csv(file) for file in files])
df['length'] = df.start_nodes + df.start_edges
df.head()
```

```python
df.dataset.unique()
```

```python
index_cmap = factor_cmap('dataset', palette=Category10[5], 
                         factors=sorted(df.dataset.unique()))
```

```python
# determine best fit line
par = np.polyfit(df.length, df.times, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
y_predicted = [slope*i + intercept  for i in df.length]
```

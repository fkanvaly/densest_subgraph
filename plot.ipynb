```python
import numpy as np
import pandas as pd

import iqplot

import bokeh.sampledata.autompg

import bokeh.io
bokeh.io.output_notebook()
```

```python
dataset = {'email-Eu-core': 26576,
           'cit-HepPh': 456124,
           'email-EuAll': 685259,
           'com-DBLP': 1366946,
           'com-Youtube': 4122514}

running_time = {"email-Eu-core":   np.arange(10), 
                "com-DBLP":      2+np.arange(10),
                "cit-HepPh":     3+np.arange(10),
                "email-EuAll":   4+np.arange(10),
                "com-Youtube": 5+np.arange(10)}

df = pd.DataFrame(running_time)
df = pd.melt(df, var_name='dataset', value_name="time")
df['size']=df.dataset.apply(lambda x: dataset[x])
df.head()
```

```python
p = iqplot.box(
    data=df,
    q="time",
    cats="size",
    q_axis="y",
    order=sorted(dataset.values())
)

bokeh.io.show(p)
```
